##**Question 1: What is Deep Learning? Briefly describe how it evolved and how it differs from traditional machine learning.**
**Answer:**
- Deep Learning is a subset of machine learning that uses artificial neural networks (ANNs) with multiple hidden layers to automatically learn representations from data. It is especially powerful for unstructured data such as images, audio, and text.
Unlike traditional ML, which often requires handcrafted features, deep learning models learn features automatically through multiple layers of abstraction.

**Evolution of Deep Learning:**
- 1940s–1960s: Early ideas of neural networks (McCulloch-Pitts neuron, Perceptron by Rosenblatt).
- 1980s–1990s: Introduction of backpropagation enabled training of multi-layer perceptrons.
- 2000s: Better hardware (GPUs) and large datasets allowed deeper networks.
- 2010s: Breakthroughs in computer vision (AlexNet, 2012), natural language processing (word embeddings, transformers).
- Today: Deep learning powers applications like self-driving cars, fraud detection, virtual assistants, medical imaging, and generative AI.

**How Deep Learning Differs from Traditional Machine Learning**
- Feature Engineering:
  - ML: Requires manual feature extraction.
  - DL: Learns hierarchical features automatically.
- Data Requirements:
  - ML: Works well with smaller datasets.
  - DL: Needs large amounts of labeled data.
- Model Complexity:
  - ML: Uses simpler algorithms (e.g., decision trees, SVM).
  - DL: Uses multi-layer neural networks with millions of parameters.
- Computation:
  - ML: Less computationally intensive.
  - DL: Requires GPUs/TPUs for large-scale training.
- Performance:
  - ML: Good for structured/tabular data.
  - DL: Superior for unstructured data (images, speech, text).

##**Question 2: Explain the basic architecture and functioning of a Perceptron. What are its limitations?**
Answer:

**Basic Architecture**
Screenshot 2025-09-21 140328.png
The Perceptron is the simplest type of artificial neural network, introduced by Frank Rosenblatt in 1958. Its main components are:
- Input Layer: Accepts input features (x₁, x₂, …, xₙ).
- Weights (wᵢ): Each input is assigned a weight that indicates its importance.
- Summation Function: Computes a weighted sum:
 z = Σ (wᵢ · xᵢ) + b
 where b is the bias.
- Activation Function: A step function that converts the sum into output (0 or 1).


**Functioning**
1. Inputs are fed into the perceptron.
2. Each input is multiplied by its weight.
3. The weighted inputs are summed along with a bias.
4. The activation function produces the final output:
   - If z ≥ θ → output = 1
   - If z < θ → output = 0
5. During training, weights are updated using the Perceptron Learning Rule to reduce classification errors.

**Limitations of Perceptron**
- Linear Separability: Works only for linearly separable problems (e.g., AND, OR), but fails on XOR.
- Limited Power: Without hidden layers, it cannot model complex or non-linear relationships.
- No Feature Hierarchy: Cannot automatically learn multiple feature levels.
- Sensitive to Scaling: Requires properly scaled inputs for better performance.
- Convergence Issues: May not converge if data is not linearly separable.

##**Question 3: Describe the purpose of activation function in neural networks. Compare Sigmoid, ReLU, and Tanh functions.**
Answer:

**Purpose**
- Activation functions introduce non-linearity into neural networks, allowing them to learn complex relationships between inputs and outputs.
- Without activation functions, the entire network would behave like a simple linear regression model, regardless of the number of layers.
- They help in controlling the output range, stabilizing training, and enabling deep networks to approximate non-linear mappings.

### Comparison of Sigmoid, ReLU, and Tanh Activation Functions

| Activation Function | Mathematical Expression | Range          | Advantages | Disadvantages | Best Used In |
|----------------------|--------------------------|----------------|------------|---------------|--------------|
| **Sigmoid** | \( f(x) = \frac{1}{1 + e^{-x}} \) | (0, 1) | - Smooth curve<br>- Interpretable as probability | - Vanishing gradient problem<br>- Outputs not zero-centered<br>- Slow convergence | Binary classification (output layer) |
| **Tanh** | \( f(x) = \frac{e^x - e^{-x}}{e^x + e^{-x}} \) | (-1, 1) | - Zero-centered<br>- Stronger gradients than sigmoid | - Still suffers from vanishing gradient<br>- Computationally expensive | Hidden layers where negative values are useful |
| **ReLU** | \( f(x) = \max(0, x) \) | [0, ∞) | - Computationally efficient<br>- Reduces vanishing gradient<br>- Sparse activations (efficient) | - Dying ReLU problem (neurons stuck at 0)<br>- Not zero-centered | Most hidden layers in deep networks (default choice) |

##**Question 4: What is the difference between Loss function and Cost function in neural networks? Provide examples.**
Answer:

### Difference Between Loss Function and Cost Function in Neural Networks

| Aspect | Loss Function | Cost Function |
|--------|---------------|----------------|
| **Definition** | A measure of error for a **single training example**. | The **average (or sum) of all loss functions** over the entire training dataset. |
| **Scope** | Instance-level (per sample). | Dataset-level (all samples). |
| **Purpose** | Tells how far the predicted output is from the actual target for one example. | Provides an overall measure of how well the model is performing on the dataset. |
| **Granularity** | Fine-grained (one observation). | Aggregated (all observations). |
| **Examples** | - Mean Squared Error (MSE) for one sample<br>- Cross-Entropy loss for one sample<br>- Hinge loss for one sample | - Mean Squared Error (MSE) averaged over all samples<br>- Categorical Cross-Entropy over dataset<br>- Binary Cross-Entropy over dataset |
| **Formula Example** | For one sample:<br> \( L(y, \hat{y}) = (y - \hat{y})^2 \) | For m samples:<br> \( J = \frac{1}{m} \sum_{i=1}^m (y_i - \hat{y}_i)^2 \) |

##**Question 5: What is the role of optimizers in neural networks? Compare Gradient Descent, Adam, and RMSprop.**
Answer:

### Role of Optimizers in Neural Networks
Optimizers are algorithms used to **update the weights and biases** of a neural network to minimize the cost (loss) function.  
They control how the model learns from data by adjusting parameters during training.  
An effective optimizer improves **training speed, convergence, and accuracy**.


### Comparison of Gradient Descent, Adam, and RMSprop

| Optimizer | Key Idea | Advantages | Disadvantages | Best Used In |
|-----------|----------|------------|---------------|--------------|
| **Gradient Descent (GD / SGD)** | Updates weights by moving in the opposite direction of the gradient of the cost function. <br> Update rule: \( w = w - \eta \cdot \nabla J(w) \) | - Simple to implement<br>- Works well for convex problems | - Slow convergence<br>- Gets stuck in local minima<br>- Requires careful learning rate tuning | Small datasets or convex optimization problems |
| **RMSprop** | Uses exponentially decaying average of squared gradients to adapt learning rates. <br> Update rule: \( w = w - \frac{\eta}{\sqrt{E[g^2] + \epsilon}} \cdot g \) | - Handles non-stationary objectives<br>- Works well in online/mini-batch learning<br>- Faster convergence than SGD | - Can sometimes converge to suboptimal points<br>- Needs hyperparameter tuning | Recurrent Neural Networks (RNNs), non-stationary problems |
| **Adam (Adaptive Moment Estimation)** | Combines **Momentum** (1st moment) and **RMSprop** (2nd moment) to adapt learning rates. <br> Update rule uses both moving averages of gradient and squared gradient. | - Fast convergence<br>- Works well with sparse gradients<br>- Less sensitive to learning rate | - Sometimes overfits<br>- Can generalize worse than SGD in some cases | Deep learning tasks (CNNs, NLP, large datasets) |

```python
#Question 6: Write a Python program to implement a single-layer perceptron from scratch using NumPy to solve the logical AND gate.
import numpy as np

X = np.array([[0, 0],
              [0, 1],
              [1, 0],
              [1, 1]])

y = np.array([[0],
              [0],
              [0],
              [1]])

np.random.seed(42)
weights = np.random.rand(2, 1)
bias = np.random.rand(1)

learning_rate = 0.1
epochs = 20

def activation(x):
    return np.where(x >= 0.5, 1, 0)

for epoch in range(epochs):
    for i in range(len(X)):
        linear_output = np.dot(X[i], weights) + bias
        y_pred = activation(linear_output)
        error = y[i] - y_pred
        weights += learning_rate * error * X[i].reshape(2, 1)
        bias += learning_rate * error

    print(f"Epoch {epoch+1}: Weights = {weights.ravel()}, Bias = {bias}")

print("\nFinal Predictions after training:")
for i in range(len(X)):
    linear_output = np.dot(X[i], weights) + bias
    y_pred = activation(linear_output)
    print(f"Input: {X[i]} → Predicted: {y_pred}, Expected: {y[i]}")
```

```python
#Question 7: Implement and visualize Sigmoid, ReLU, and Tanh activation functions using Matplotlib.
import numpy as np
import matplotlib.pyplot as plt

# Input range
x = np.linspace(-10, 10, 100)

# Activation functions
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def relu(x):
    return np.maximum(0, x)

def tanh(x):
    return np.tanh(x)

# Compute outputs
y_sigmoid = sigmoid(x)
y_relu = relu(x)
y_tanh = tanh(x)

# Plotting
plt.figure(figsize=(15, 4))

plt.subplot(1, 3, 1)
plt.plot(x, y_sigmoid, 'b')
plt.title("Sigmoid Function")
plt.grid(True)

plt.subplot(1, 3, 2)
plt.plot(x, y_relu, 'g')
plt.title("ReLU Function")
plt.grid(True)

plt.subplot(1, 3, 3)
plt.plot(x, y_tanh, 'r')
plt.title("Tanh Function")
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
#Question 8: Use Keras to build and train a simple multilayer neural network on the MNIST digits dataset. Print the training accuracy.
import tensorflow as tf
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.utils import to_categorical

# Load MNIST dataset
(x_train, y_train), (x_test, y_test) = mnist.load_data()

# Normalize data
x_train = x_train / 255.0
x_test = x_test / 255.0

# One-hot encode labels
y_train = to_categorical(y_train, 10)
y_test = to_categorical(y_test, 10)

# Build model
model = Sequential([
    Flatten(input_shape=(28, 28)),
    Dense(128, activation='relu'),
    Dense(64, activation='relu'),
    Dense(10, activation='softmax')
])

# Compile model
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

# Train model
history = model.fit(x_train, y_train, epochs=10, batch_size=128, verbose=1)

# Print final training accuracy
train_acc = history.history['accuracy'][-1]
print(f"\nFinal Training Accuracy: {train_acc:.4f}")
```

```python
#Question 9: Visualize the loss and accuracy curves for a neural network model trained on the Fashion MNIST dataset. Interpret the training behavior.
import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.datasets import fashion_mnist
import matplotlib.pyplot as plt


(x_train, y_train), (x_test, y_test) = fashion_mnist.load_data()


x_train, x_test = x_train / 255.0, x_test / 255.0


model = models.Sequential([
    layers.Flatten(input_shape=(28, 28)),
    layers.Dense(256, activation='relu'),
    layers.Dense(128, activation='relu'),
    layers.Dense(10, activation='softmax')
])


model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])


history = model.fit(x_train, y_train, validation_split=0.2, epochs=12, batch_size=64, verbose=1)


fig, axs = plt.subplots(1, 2, figsize=(12, 5))


axs[0].plot(history.history['loss'], label='Train')
axs[0].plot(history.history['val_loss'], label='Validation')
axs[0].set_title('Loss Curve')
axs[0].set_xlabel('Epochs')
axs[0].set_ylabel('Loss')
axs[0].legend()


axs[1].plot(history.history['accuracy'], label='Train')
axs[1].plot(history.history['val_accuracy'], label='Validation')
axs[1].set_title('Accuracy Curve')
axs[1].set_xlabel('Epochs')
axs[1].set_ylabel('Accuracy')
axs[1].legend()


plt.show()
```

##Question 10: You are working on a project for a bank that wants to automatically detect fraudulent transactions. The dataset is large, imbalanced, and contains structured features like transaction amount, merchant ID, and customer location. The goal is to classify each transaction as fraudulent or legitimate.
Explain your real-time data science workflow:
- How would you design a deep learning model (perceptron or multilayer NN)?
- Which activation function and loss function would you use, and why?
- How would you train and evaluate the model, considering class imbalance?
- Which optimizer would be suitable, and how would you prevent overfitting?

**Answer:**

**1. Model Design**
- Type of Model: Multilayer Neural Network (MLP) is preferable over a single-layer perceptron due to the complex, non-linear patterns in fraud detection.
- Input Layer: Number of neurons = number of features (e.g., transaction amount, merchant ID, customer location, etc.).
- Hidden Layers: 2–3 hidden layers with 64–128 neurons each (tune based on dataset size and complexity).
- Output Layer: Single neuron with sigmoid activation to predict the probability of fraud.
**2. Activation and Loss Functions**

| Component | Choice | Reason |
|-----------|--------|--------|
| Hidden Layer Activation | ReLU (or LeakyReLU) | - Fast computation<br>- Sparse activations<br>- Avoids vanishing gradient problem<br>- LeakyReLU prevents “dying ReLU” issue |
| Output Layer Activation | Sigmoid | - Suitable for binary classification<br>- Outputs probability between 0 and 1 |
| Loss Function | Binary Cross-Entropy (BCE) with class weights <br>or Focal Loss | - BCE is standard for binary tasks<br>- Class weights handle severe class imbalance<br>- Focal loss focuses training on rare/fraudulent examples, reducing the effect of easy negatives |

**3. Training and Evaluation (Handling Class Imbalance)**
- Data Preprocessing:
   - Normalize numerical features (e.g., transaction amount).
   - One-hot encode categorical features (merchant ID, location, etc.).
- Handling Imbalance:
   - Use class weights in the loss function to penalize misclassification of minority class (fraud).
   - Alternatively, oversample fraudulent transactions (SMOTE) or undersample legitimate transactions.
- Evaluation Metrics: Accuracy is misleading; prefer:
   - Precision, Recall, F1-Score
   - ROC-AUC (measures ability to distinguish fraud vs. legitimate transactions)
   - PR-AUC (Precision-Recall curve, especially for highly imbalanced data)
**4. Optimizer and Overfitting Prevention**
- Optimizer: Adam
   - Adaptive learning rates for each parameter
   - Fast convergence for large datasets
- Overfitting Prevention:
   - Dropout: Randomly drop neurons during training (e.g., 20–50%)
   - Early Stopping: Stop training when validation loss stops improving
   - Regularization: L2 weight decay to penalize large weights
   - Batch Normalization: Stabilizes training and improves generalization
**5. Training Workflow**
   1. Split dataset into train, validation, and test sets (e.g., 70-15-15).
   2. Apply feature scaling and encoding.
   3. Define the MLP architecture (input → hidden → output).
   4. Assign class weights to handle imbalance.
   5. Compile the model using Adam optimizer and binary cross-entropy loss.
   6. Train with mini-batches, monitor validation metrics.
   7. Evaluate on the test set using F1-score, ROC-AUC, and PR-AUC.
   8. Fine-tune hyperparameters (learning rate, hidden layers, dropout rate) for optimal performance.
